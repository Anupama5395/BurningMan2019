```
import numpy as np
# import audaugio
wav_dir=os.path.join(project_dir,'wav')
output_dir=os.path.join(project_dir,'wav_aug')
wav_files_list=os.listdir(wav_dir)
wav_files_list[0:5]
```

```
import numpy as np
from tqdm import tqdm_notebook
from scipy.io import wavfile
for file_ in tqdm_notebook(wav_files_list):
  fs, data_in = wavfile.read(os.path.join(wav_dir,file_))
  for n_l in [0,300,500]:
    data_noisy=data_in+n_l*np.random.randn(len(data_in))
    data_noisy=data_noisy.astype('int16')
    file_out=file_.strip('.wav')+'_'+str(n_l)+'.wav'
    wavfile.write(os.path.join(output_dir,file_out),fs,data_noisy)
```
